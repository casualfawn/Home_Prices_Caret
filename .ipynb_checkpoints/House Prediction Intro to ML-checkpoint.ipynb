```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats

from sklearn.preprocessing import StandardScaler

from pycaret.regression import setup, compare_models

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import KFold, cross_val_score

from sklearn.linear_model import Lasso, ElasticNet, LassoLars
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import BayesianRidge
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
pd.set_option('display.max_columns', None)
```

```python
train_data = pd.read_csv('train.csv', index_col='Id')
test_data = pd.read_csv('test.csv', index_col='Id')
```

```python
print(len(train_data))
print(len(test_data))
```

```python
data0 = pd.concat([train_data, test_data]).reset_index(drop=True)
```

```python
data0.head()
```

```python
data0.tail()
```

```python
target = train_data['SalePrice']
```

```python
data1 = data0.copy()
data1 = data1.drop('SalePrice', axis=1)
```

```python
data1.MSSubClass = data1.MSSubClass.astype('str')
data1.MoSold = data1.MoSold.astype('str')
```

```python
data1.describe()
```

```python
data2 = data1.copy()
```

# Categorical data cleaning

```python
categorical_cols = [col for col in data2.columns if data2[col].dtype == 'object']

for col in data2[categorical_cols].loc[: ,data2[categorical_cols].isnull().sum()>0].columns:
    if col in ['Alley', 'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'FireplaceQu', 
               'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond', 'PoolQC', 'Fence', 'MiscFeature']:
        data2[col] = data2[col].fillna('None')
    else:
        data2[col] = data2[col].fillna(data2[col].mode()[0])
```

```python
data3 = data2.copy()
```

# Numerical data cleaning

```python
numerical_cols = [col for col in data3.columns if data3[col].dtype == 'float64' or data3[col].dtype == 'int64']
```

```python
data3.LotFrontage = data3.LotFrontage.fillna(data3.LotFrontage.median())
data3.GarageYrBlt = data3.GarageYrBlt.fillna(data3['YearBuilt'])
for col in ['MasVnrArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF', 
            'GarageCars', 'GarageArea', 'BsmtFullBath', 'BsmtHalfBath']:
    data3[col] = data3[col].fillna(0)
```

```python
data3.isnull().sum().sum()
```

```python
data4 = data3.copy()
```

# Feature Engineering

```python
data4['TotalBath'] = data4.BsmtFullBath + data4.FullBath + .5*data4.BsmtHalfBath + .5*data4.HalfBath
data4['Area'] = data4.GrLivArea + data4.TotalBsmtSF
data4['HouseAge'] = data4.YrSold-data4.YearBuilt
data4['Total_Home_Quality'] = data4['OverallQual'] + data4['OverallCond']
```

```python
data5 = data4.copy()
```

# Feature Transformation

```python
for col in data5[categorical_cols]:
    dummies = pd.get_dummies(data5[col], prefix=col, drop_first=True)
    data5 = pd.concat([data5, dummies], axis='columns')
    data5 = data5.drop([col], axis='columns')
```

```python
data6 = data5.copy()
```

```python
skew_cols = []
for col in data6[numerical_cols]:
    if scipy.stats.skew(data6[col])>=5:
        print(col, 'skewness:', scipy.stats.skew(data6[col]))
        skew_cols.append(col)
```

```python
for col in skew_cols:
    data6[col] = np.log1p(data6[col])
```

```python
data7 = data6.copy()
```

# Feature Scaling

```python
scaler = StandardScaler()

scaler.fit(data7)
data7 = pd.DataFrame(scaler.transform(data7), index=data7.index, columns=data7.columns)
```

```python
data8 = data7.copy()
```

# Target Transformation

```python
print('Sale Price mean: ${}'.format(int(round(target.mean(),0))))
print('Sale Price median: ${}'.format(int(round(target.median()))))
print('Sale Price min: ${}'.format(int(round(target.min()))))
print('Sale Price max: ${}'.format(int(round(target.max()))))

sns.set(rc={'figure.figsize':(15,10)})
sns.histplot(target)
plt.show()
```

```python
sns.histplot(np.log(target))
plt.show()
```

```python
log_target = np.log(target)
```

# Split back to train and test datasets

```python
train = data8.loc[:1460-1,:]
test = data8.loc[1460:,:]
```

# Compare models

```python
_ = setup(data=pd.concat([train, log_target], axis=1), target='SalePrice')
```

```python
compare_models()
```

# Model Selection

```python
models = {
    'lasso': Lasso(),
    'br': BayesianRidge(),
    'ada' : AdaBoostRegressor(),
    'gbr': GradientBoostingRegressor(verbose=0),
    'rf': RandomForestRegressor()
}
```

```python
for name, model in models.items():
    model.fit(train, log_target)
    print(name + " trained")
```

```python
results = {}

kf = KFold(n_splits=10)

for name, model in models.items():
    result = np.exp(np.sqrt(-cross_val_score(model, train, log_target, scoring='neg_mean_squared_error', cv=kf)))
    results[name] = result
```

```python
for name, result in results.items():
    print("----------\n" + name)
    print(np.mean(result))
    print(np.std(result))
```

```python
predictions = (
    0.1 * np.exp(models['lasso'].predict(test)) +
    0.2 * np.exp(models['br'].predict(test)) +
    0.2 * np.exp(models['ada'].predict(test)) +
    0.3 * np.exp(models['gbr'].predict(test)) +
    0.2 * np.exp(models['rf'].predict(test))
)
```

```python
predictions
```

```python
submission = pd.concat([pd.Series(test_data.index), pd.Series(predictions, name='SalePrice')], axis=1)
submission
```

```python
submission.to_csv('submission_feature_engineering.csv', index=False)#, header=True)
```
